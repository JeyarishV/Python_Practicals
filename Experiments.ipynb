```python
pro_table = {
    "E": 1.51, "M": 1.45, "A": 1.42, "L": 1.21, "K": 1.16, "F": 1.13, "Q": 1.11, "W": 1.08,
    "I": 1.08, "V": 1.06, "D": 1.01, "H": 1.00, "R": 0.96, "T": 0.83, "S": 0.77, "C": 0.70,
    "Y": 0.69, "N": 0.67, "P": 0.57, "G": 0.57
}

def alpha_helix(sequence, position):
    # TASK 1: SLICING SEQ INTO FRAMES OF SIX
    frame1 = sequence[position:position + 6]
    print("FRAME1 =", frame1)

    # TASK 2: REFER TO DICT AND CHECK PROPENSITY
    frame2 = ""
    for x in frame1:
        if pro_table[x] >= 1:
            frame2 += x 
    print("FRAME2 =", frame2)

    # CALCULATE AVG PROPENSITY IF 4 AMINO ACIDS ARE IN FRAME 2
    if len(frame2) >= 4:
        avg_propensity = sum(pro_table[x] for x in frame1) / len(frame1)
        print("This is a alpha helix nucleation site with propensity ", avg_propensity , "at position", position)

seq = input("Enter a protein sequence: ")
position = 0
for x in seq:
    while position<len(seq):
        alpha_helix(seq,position)
        position+=6
```

```python
def fasta_reader(file_path):
    sequences = {}
    with open(file_path, 'r') as file:
        lines = file.readlines()
        key = None
        value = ''
        for line in lines:
            line = line.rstrip()  # Remove trailing whitespace characters
            if line.startswith('>'):
                if key is not None:
                    sequences[key] = value
                key = line
                value = ''
            else:
                value += line
        if key is not None:
            sequences[key] = value
    return sequences

file_path = "C:\\Users\\jeyar\\Downloads\\Covid-spike.txt"
output_dict = fasta_reader(file_path)
#print(output_dict)
#To make the dictionary more readable
for key, value in output_dict.items():
    print(f"{key}\n{value}\n")
```
